#### string slicing

```python
a = "0123456789"

print(a[2:5]) #starts with 2 index ends at 5-1 index

print(a[:6]) #slice from the start

print(a[2:])#slice till the end

print(a[-9:-2]) # starts from last -1 os 8
```

## String Modification

```python
a = "     a B c D e F g H i J k L m N o P q R s T u V w X y Z     "

print(a)

Upper = print(a.upper())

Lower = print(a.lower())

NoSpace = print(a.strip()) #remove space from start and end
```

### String Replace

```python
a = "Hello!"
print(a.replace("Hell", "bol"))
```

## Split string


```python
a = "a , b"

b = "0123456"

print(a.split(","))

print(b.split("3"))
```
