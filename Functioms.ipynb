# we can also satisfy a value for particular keyword while calling a function

```python
def my_name(name = "Aniket", Animal = "Dog"):
    print("My Name is : ", name, " My fav animal is : ", Animal)

my_name()
my_name("idk","idk2")
my_name(name = "Sagar")
my_name(Animal = "Cat")
my_name(name = "DEv", Animal = "Rat")
my_name(Animal = "IDK", name = "Nilesh")
```

# Recursion

```python
def countdown(n):
  if n <= 0:
    print("Done!")
  else:
    print(n)
    countdown(n - 1)

countdown(5)
```

```python
def factorial(n):
  # Base case
  if n == 0 or n == 1:
    return 1
  # Recursive case
  else:
    return n * factorial(n - 1)

print(factorial(5))
```

# setting a limit on recursion
