# init

```python
class init():
    def __init__(main,a,b):
        main.a = a
        main.b = b

p1 = init("sagar","prajapati")

print(p1.a)
```

# Calling in another def

```python
class init():
    def __init__(main,a,b):
        main.a = a
        main.b = b

    def greet(main):
        print("hello the value is ",main.a)

p1 = init("Aniket","sagagr")

p1.greet()
```

# inheritance

```python
class parent():
    def __init__(main,a,b):
        main.a = a
        main.b = b

    def printt(main):
        print(main.a,main.b)

class child(parent):
    pass

p1 = child("Mukesh","aniket")
p1.printt()
```
