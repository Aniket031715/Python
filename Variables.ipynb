## Global & Local Variables

```python
x = "This is a Global Variable"

def myfunc():
    x = "This is a Local Variable"
    print(x)

myfunc()

print(x)
```

### Global function

#### when this function is used the global variables values can be changed from local variable

```python
x = "This is a Global Variable"

def myfunc():
    global x
    x = "This is a Local Variable"
    print(x)

myfunc()

print(x)
```
