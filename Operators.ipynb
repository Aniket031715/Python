# Operators

## Booleans (checking if a statement is true or false)

```python
print(10>9)
```

```python
print(10 == 10)
```

```python
print(9 == 6)
```

## Arithmetic operators

```python
print(2 + 3)
print(5 - 2)
print(2 * 5)
print(2 ** 5) #equal to 2 ^ 5
print(5 / 2)
print(5 // 2) #rounds the answer
print(15 // 2)
print(15 % 2) #remainder of devision
```

## Assignment Operator

```python
x = 5
print(x)

x += 5
print(x)

x -= 5
print(x)

x *= 5
print(x)

x /= 5
print(x)
```

## Ternary Operator

```python
x = 5

day = "Weekday!" if x <= 5 else "WEEKEND YEPEEEEEEE"

print(day)
```

```python
num = 7

x = "Fri" if num == 5 else "Sat" if num == 6 else "Sun" if num == 7 else "weekday"

print(x)
```

## Comparision Operator

```python
x = 6
y = 5

print(x==y)
print(x!=y)
print(x>y)
print(x<y)
print(x>=y)
print(x<=y)
```

```python
x = 5

print(1 < x < 10)
```

## Logical Operator

```python
x = 5
y = 10

print(x > 1 and y < 15)

print(x > 1 or x < 2)

print(not(x > 1 and y < 15))
```

## Identity Operator

```python
x = 5
y = 5

print(x is y)
print(x is not y)
```

## Membership Operator

```python
a = ["X","Y","Z"]

print("x" in a)

print("X" in a)

print("y" not in a)
```

```python
string = "Bolo Bhai!!"

print("B" in string)

print("H" not in string)
```

## Bitwise Operator

```python
x = 1
y = 1
```

```python
#AND

print(x & y)
```

```python
#OR

print(x | y)
```

```python
#XOR

print(x ^ y)

x = 0
y = 1

print(x ^ y)

x = 0
y = 0

print(x ^ y)
```
