# IF-Else

```python
a = 33
b = 200
if b > a:
  print("b is greater than a")
```

```python
age = 20
if age >= 18:
  print("You are an adult")
  print("You can vote")
  print("You have full legal rights")
```

```python
is_logged_in = True
if is_logged_in:
  print("Welcome back!")
```

```python
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

```python
score = 75

if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
elif score >= 60:
  print("Grade: D")
```

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

```python
temperature = 22

if temperature > 30:
  print("It's hot outside!")
elif temperature > 20:
  print("It's warm outside")
elif temperature > 10:
  print("It's cool outside")
else:
  print("It's cold outside!")
```

```python
a = 2
b = 330
print("A") if a > b else print("B")
```

```python
a = 10
b = 20
bigger = a if a > b else b
print("Bigger is", bigger)
```

```python
x = 15
y = 20
max_value = x if x > y else y
print("Maximum value:", max_value)
```

```python
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True")
```

```python
temperature = 25
is_raining = False
is_weekend = True

if (temperature > 20 and not is_raining) or is_weekend:
  print("Great day for outdoor activities!")
```

```python
age = 25
has_license = True

if age >= 18:
  if has_license:
    print("You can drive")
  else:
    print("You need a license")
else:
  print("You are too young to drive")
```

## Match

```python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
```

```python
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
```

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

# While

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```
