## Copy A LIST

```python
list1 = ['H','E','Y']

print(list1)

list2 = list1.copy()
list3 = list(list1)

print("This is List2", list2)
print("This is list3", list3)
```

## Join a list

```python
list1 = ['H','E','Y']

print(list1)

list2 = ['T','H','E','R','E']

print(list2)

list3 = list1 + list2
print(list3)

list1.extend(list2)
print(list1)

for x in list3 :
    list1.append(x)

print(list1)
```

# Dictionary
- ordered
- changeable
- no duplicates

```python
CarsDict = {
    "brand": "audi",
    "model": "r4",
    "year": 1994
}

print(CarsDict)

print(CarsDict["year"])
```

```python
print(len(CarsDict))
```

```python
ListDict = {
    "Colors" : ['red','white','blue']
}

print(ListDict)
```

```python
x = ListDict["Colors"]

print(x)

y = CarsDict["year"]
print(y)

y = y/2
print(y)

x = CarsDict.keys()
print(x)

y = CarsDict.values()
print(y)

z = CarsDict.items()
print(z)
```

```python
CarsDict["year"] = 2000

print(CarsDict)

CarsDict["Speed"] = 200
print(CarsDict)

CarsDict.pop("Speed")
print(CarsDict)

CarsDict.clear()
print(CarsDict)
```

## Nested Dictionary

```python
MyFamily = {
    "Me" : {
        "Name" : "Aniket",
        "Age" : "20"
    },
    "Father" : {
        "Name" : "Mukesh",
        "Age" : "49"
    },
    "Mother" : {
        "Name" : "Sushila",
        "Age" : "48"
    },
    "Car" : CarsDict
}

print(MyFamily)
```
